```python
import pandas as pd

gps=pd.read_csv('googleplaystore.csv')
gps.head()
```

```python
gps.info()
```

```python
gps.describe()
gps['Rating'].value_counts()
```

```python
#Fill not null data with below conditions
gps['Rating']=gps['Rating'].fillna(gps['Rating'].mean())
v_cst=gps['Type'].value_counts()
print(v_cst)

v_csc=gps['Content Rating'].value_counts()
print(v_csc)
```

```python
gps['Type']=gps['Type'].fillna('Free')
gps['Content Rating']=gps['Content Rating'].fillna('Everyone')
gps.head()
```

```python
gps.info()
```

```python
gps=gps.sort_values(by=['Size'])
gps.head()
```

```python
gps=gps.drop(['Current Ver','Android Ver'], axis=1)
gps.head()
```

```python
gps['Rating_round']=gps['Rating'].round()
gps.head()
```

```python
def get_rating_cat(rate):
    cat=''
    if rate==1: cat='Very Bad'
    elif rate==2: cat='Bad'
    elif rate==3: cat='So So'
    elif rate==4: cat='Good'
    elif rate==5: cat='Very Good'
    else : cat='ex'
        
    return cat

gps['Rating_cat']=gps['Rating_round'].apply(lambda x: get_rating_cat(x))
gps.head()
```

```python
gps=gps.drop(10472,axis=0)
```

```python
cat_rate_mean = {'Rating':'mean'}
gps.groupby('Category').agg(cat_rate_mean)
```
